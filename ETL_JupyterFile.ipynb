```python
#Dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests
import psycopg2
from bs4 import BeautifulSoup
```

```python
#Reading CSV
Baseball = pd.read_csv ('Teams.csv')
Baseball.head()
```

```python
BB = Baseball.dropna(subset=['LgWin'])
BB.head()
```

```python
WSParticipants= BB[BB.LgWin != 'N']
WSParticipants.head()
```

```python
RecentWSParticipants = WSParticipants[-92:]
RecentWSParticipants.head()
```

```python
WorldSeriesTeams = RecentWSParticipants.drop(['franchID','divID','Ghome','teamIDBR','teamIDlahman45','teamIDretro'], axis=1)
WorldSeriesTeams[["BBA", "HA","IPouts"]] = WorldSeriesTeams[["BBA", "HA","IPouts"]].astype('float64')
WorldSeriesTeams['IP']=WorldSeriesTeams['IPouts']/3
WorldSeriesTeams['WHIP'] = WorldSeriesTeams['BBA'] + WorldSeriesTeams['HA']
WorldSeriesTeams['WHIP'] = WorldSeriesTeams['WHIP'] / WorldSeriesTeams['IP']
```

```python
WSeriesTeams = WorldSeriesTeams[['yearID', 'lgID', 'teamID', 'name','G', 'W', 'L', 'DivWin','WCWin','LgWin','WSWin','R','AB','H','2B','3B','HR','BB','SO','SB','CS','HBP','SF','IP','RA','ER','ERA','WHIP','CG','SHO','SV','HA','HRA','BBA','SOA','E','DP','FP','park','attendance', 'BPF', 'PPF']]
```

```python
#Cleaned CSV File
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', -1)
WorldSeriesFinal = WSeriesTeams.reset_index()
WorldSeriesFinal
```

```python
#Website we used to scrape
url = 'https://www.baseball-almanac.com/ws/wstv.shtml'
```

```python
#Finished Scraping DataFrame
tables = pd.read_html(url)
tables_unclean = tables[6]
df_unclean = pd.DataFrame(tables_unclean)
df_unclean = df_unclean.drop([0,55,56], axis=0)
df_unclean.columns = df_unclean.iloc[0]
df_clean = df_unclean[1:]
df_clean.reset_index(inplace = True, drop = True)
df_clean
```

```python
#Renaming the year column to yearID since we are merging on that column
df_clean = df_clean.rename(columns={'Year / Series': 'yearID'})
df_clean
```

```python
df_clean["yearID"] = WorldSeriesFinal["yearID"].astype(int)
```

```python
#Merging both dataframes on the "yearID" column
merge_dataframe = pd.merge(df_clean, WorldSeriesFinal, on="yearID")
merge_dataframe
```

```python
#Uploading the final to PostgresSQl
#First, we create the database name 'production' in Postgres
#Then, we link our merged dataframe with the engine 

import sqlalchemy
from sqlalchemy import create_engine 
engine = sqlalchemy.create_engine('postgresql://postgres:postgres@localhost/production')
merge_dataframe.to_sql('merge_dataframe', engine)
```
